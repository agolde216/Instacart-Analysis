```python
df = ords_prods_merged[:1000000]
```

```python
# define function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#apply function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
#function for busiest day
result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
ords_prods_merged['busiest_day'] = result
```

```python
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

```python
#function for 2 busiest days
result1 = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result1.append("Busiest day")
  elif value == 1:
    result1.append("2nd busiest day")
  elif value == 3:
    result1.append("2nd slowest day")
  elif value == 4:
    result1.append("Least busy")
  else:
    result1.append("Regularly busy")
```

```python
result1
```

```python
ords_prods_merged['busiest_days'] = result1
```

```python
ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

```python
ords_prods_merged['hour_of_day_ordered'].value_counts(dropna = False)
```

```python
#function for average hours ordered
result2 = []

for value in ords_prods_merged["hour_of_day_ordered"]:
  if value == 10:
    result2.append("Most orders")
  elif value == 3:
    result2.append("Fewest orders")
  else:
    result2.append("Average orders")
```

```python
result2
```

```python
ords_prods_merged['busiest_period_of_day'] = result2
```
